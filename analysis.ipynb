```python
print(df.describe())
df['Order Date']=pd.to_datetime(df['Order Date'],format='mixed',dayfirst=False)
df['Ship Date']=pd.to_datetime(df['Ship Date'],format='mixed',dayfirst=False)


```

```python
df['order Year']=df['Order Date'].dt.year
df['order Month']=df['Order Date'].dt.month
df['order month name']=df['Order Date'].dt.strftime('%b')
df['Ship Days']=(df['Ship Date']-df['Order Date']).dt.days
```

```python
df['Is Loss']=df['Profit']<0
```

```python
print("date types after fix:")
print(df[['Order Date','Ship Date']].dtypes)
```

```python
print("\nShipping Days sample:")
print(df['Ship Days'].describe())
```

```python
print("\nLoss transactions:")
print(df['Is Loss'].value_counts())
```

```python
print("\nyears in dataset:")
print(df['order Year'].value_counts().sort_index())
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
yearly=df.groupby('order Year')[['Sales','Profit']].sum().reset_index()
print("yearly sales&Profit:")
print(yearly)
```

```python
category=df.groupby('Category')[['Sales','Profit']].sum().reset_index()
print("\nSales by category:")
print(category)
```

```python
subcat=df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
print("\n top 10 sub-categories:")
print(subcat)
```

```python
region=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("\nSales by region:")
print(region)
```

```python
plt.figure(figsize=(8,4))
plt.plot(yearly['order Year'], yearly['Sales'], marker='o', color='steelblue', linewidth=2)
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.savefig('yearly_sales.png')
plt.show()
print("Chart saved as yearly_sales.png")



```

```python
plt.figure(figsize=(7,4))
sns.barplot(data=category,x='Category',y='Profit',hue='Category',palette='Greens_d',legend=False)
plt.title('Profit by Category')
plt.tight_layout()
plt.savefig('Profit_by_category.png')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.barplot(data=category, x='Category', y='Sales', hue='Category',palette='Blues_d',legend=False)
plt.title('Sales by Category')
plt.tight_layout()
plt.savefig('Sales_by_category.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
subcat.sort_values().plot(kind='barh',color='Steelblue')
plt.title('top 10 sub_categories by sales')
plt.xlabel('Total Sales')
plt.tight_layout()
plt.savefig('top_subcategoried.png')
plt.show()
```

```python
plt.figure(figsize=(6,6))
region.plot(kind='pie',autopct='%1.1f%%',startangle=90,colors=['#4C72B0','#55A868','#C44E52','#8172B2'])
plt.title('sales by region')
plt.ylabel('')
plt.tight_layout()
plt.savefig('sales_by_region.png')
plt.show()
```
